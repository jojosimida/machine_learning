```python
import pandas as pd
import numpy as np
import nltk
import pickle
```

```python
train_df = pd.read_csv('D:/Program/dataset/Spooky_Author_Identification/train.csv')
test_df = pd.read_csv('D:/Program/dataset/Spooky_Author_Identification/test.csv')

train_df.head()
```

```python
texts=[]
for text in train_df['text']:
    texts.append(text)
    
test_texts=[]
for text in test_df['text']:
    test_texts.append(text)
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
# Embedding
max_features = 20000
maxlen = 5000

tokenizer = Tokenizer(num_words=max_features)
tokenizer.fit_on_texts(texts+test_texts)
```

```python
sequences = []
# 按输入的文本列表生成每个文本的one-hot序列
for seq in tokenizer.texts_to_sequences_generator(texts):
    sequences.append(seq)
#     print(seq)
```

```python
tokenizer.fit_on_sequences(sequences)
```

```python
matrix = tokenizer.texts_to_matrix(texts, mode='tfidf')
matrix = pad_sequences(matrix, maxlen=maxlen,padding='post', truncating='post')
print(np.array(matrix).shape)
```

```python
test_matrix = tokenizer.texts_to_matrix(test_texts, mode='tfidf')
test_matrix = pad_sequences(test_matrix, maxlen=maxlen,padding='post', truncating='post')
print(np.array(test_matrix).shape)
```

```python
with open('keras_input_train.pkl', 'wb') as f:
    pickle.dump(matrix, f)
with open('keras_input_test.pkl', 'wb') as f:
    pickle.dump(test_matrix, f)

```
